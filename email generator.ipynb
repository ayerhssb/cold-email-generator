```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    groq_api_key = '',
    temperature=0,
    # max_tokens=None,
    # reasoning_format="parsed",
    # timeout=None,
    # max_retries=2,
    # other params...
)
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://jobs.cisco.com/jobs/IsAJob?projectId=1449112&tags=2027summerinternship")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTIONS:
    The scraped text is from a career page of a website.
    Your job is to extract the job postings and return them in a JSON format containing the following keys:
    `role`, `experience`, `skills`, and `description`.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE
    """
)
print(prompt_extract)
```

```python
chain_extract = prompt_extract | llm #it is for langchain chain basically...you're getting  prompt and you are passing it to llm
res= chain_extract.invoke(input = {'page_data':page_data})
print(res.content)
type(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
print(json_res)
```

```python
#iterate the dataframe we've made and insert into chromadb
#if used only Client it will make chroma DB in memory....but in PersistentClient it will make chroma DB in disk
import uuid
import chromadb

client=chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(
            documents=row["Techstack"],
            metadatas={"links":row["Links"]},
            ids=[str(uuid.uuid4())])
    
```

```python
links=collection.query(query_texts = ["Experience in Python","Experience in React"], n_results=2).get('metadatas')
# collection.get('metadata')
print(links)
```
